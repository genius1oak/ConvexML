## 2. Discover a real truth yourself

The safest, most satisfying early wins are **things you find in the data** — un-leakable, and they *are* the core lesson. Run the three cells below. Each is ~10 lines of pandas and each overturns a common SEO belief.

Every number is **computed live from the shipped CSV** — nothing is hardcoded.

```python
import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(df.shape[0], "rows,", df.shape[1], "columns")
df.head(3)
```

### Discovery A — "High search volume means more traffic." Does it?

```python
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")
print("Near zero -> keyword search volume barely predicts the traffic a page actually gets.")
```

### Discovery B — the CTR cliff by position
Click-through rate is not flat: it collapses as you move down the results.

```python
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos.round(4).to_string())
ctr_by_pos.plot(kind="bar", title="Mean CTR by position tier (impressions >= 100)", ylabel="CTR");
```

### Discovery C — is longer content the lever?
Compare word count for **declining** vs **growing** pages.

```python
wc = df.groupby("trend_direction")["word_count"].median()
print(wc.round(0).to_string())
print("\n'down' vs 'up' pages have almost the same median word count -> length is not the lever.")
```

### Discovery D — Does `content_age_days` relate to `trend_direction`?

```python
trend_mapping = {'down': -1, 'stable': 0, 'up': 1}
df['trend_numerical'] = df['trend_direction'].map(trend_mapping)

corr_age_trend = df['content_age_days'].corr(df['trend_numerical'])
print(f"Correlation between content_age_days and trend_direction: {corr_age_trend:.3f}")
print("A slightly negative correlation suggests that older content is marginally more likely to be 'down' trending, but the relationship is still very weak.")
```

```python
filtered_df = df[df['impressions_90d'] >=100]
ctr_by_pos_filtered =filtered_df.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos_filtered.round(4).to_string())
ctr_by_pos_filtered.plot(kind="bar", title="Mean CTR by position tier(impressions >= 100)", ylabel="CTR");
```

```python
ctr_by_pos_content = filtered_df.groupby(['position_tier','content_type'])['ctr'].mean()
print(ctr_by_pos_content)
```

```python
ctr_by_pos_content = filtered_df.groupby(['position_tier','content_type'])['ctr'].mean()
worst_ctr_content_type = ctr_by_pos_content.unstack().idxmin(axis=1)
print("Content type with worst CTR at each position tier:")
print(worst_ctr_content_type.to_string())
```

This output shows, for each `position_tier`, which `content_type` had the lowest average Click-Through Rate. This can be insightful for understanding which content types might be underperforming in certain search positions.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Calculate the actual CTR values for the worst performing content types
worst_ctr_values = pd.Series(dtype=float)
for tier, content_type_name in worst_ctr_content_type.items():
    worst_ctr_values.loc[tier] = ctr_by_pos_content.loc[(tier, content_type_name)]

plt.figure(figsize=(12, 7))
sns.barplot(x=worst_ctr_values.index, y=worst_ctr_values.values, hue=worst_ctr_values.index, palette='viridis', legend=False)
plt.title('CTR Value of Worst Performing Content Type by Position Tier')
plt.xlabel('Position Tier')
plt.ylabel('Worst CTR Value')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
